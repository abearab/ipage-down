```python
import os 
import re
import glob
import pandas as pd 
import ipage_down as ipd
```

```python
def gmt2page_index(PATH):
    """
    """
    with open(PATH) as raw:
        lines = [line for line in raw.read().splitlines()]
        lines = [line.split('\t') for line in lines] # if re.search(gs, line)]

        gsets = [line[0] for line in lines]
        genes = [line[2:] for line in lines]
        
        gs2gene = [(gn,gs) for gn,gs in zip(gsets,genes)]
        
        allgenes = {gene for x in genes for gene in x}
        
        out = dict()
        for gene in allgenes: # slow!!
            out[gene] = [gs for gs,gn in zip(gsets,genes) if gene in gn]
        
    return out



def gmt2page_names(PATH):
    """
    """

    with open(PATH) as raw:
        lines = [line for line in raw.read().splitlines()]
        lines = [line.split('\t') for line in lines] # if re.search(gs, line)]

        gsets = [line[0] for line in lines]
        cards = [line[1] for line in lines]
        
        out = pd.DataFrame([(gs,gs.replace('_',' '),card) for gs,card in zip(gsets,cards)])
        
    return out
```

```python
%%time
gmts = glob.glob('msigdb_v7.4/*/*.gmt')
gmts.sort()

for gmt in gmts:
    msig_id = gmt.split('/')[0]+'_'+gmt.split('/')[1]
    direcory = 'annotations/'+msig_id
    index_file = direcory+'/'+msig_id+'_index.txt.gz'
    names_file = direcory+'/'+msig_id+'_names.txt.gz'    
    
    if not os.path.exists(direcory):
        os.mkdir(direcory)

    print (msig_id)
    if not os.path.exists(index_file):
        index = gmt2page_index(gmt)
        ipd.write_page_index(index, index_file)
        print ('index is done!')
    
        del index
    
    if not os.path.exists(names_file):
        names = gmt2page_names(gmt)
        ipd.write_page_names(names,names_file)
        print ('name is done!')
        del names
        
    print ('\n\n')
```
